```python
# loáº¡i bá» cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t, dáº¥u cÃ¢u
import string
map_table = str.maketrans("", "", string.punctuation)
all_text = all_text.translate(map_table)
print(all_text)
    
```

```python
# loáº¡i bá» cÃ¡c stopword trong all_text
stop_word = {"bá»‹","bá»Ÿi","cáº£","cÃ¡c","cÃ¡i","cáº§n","cÃ ng","chá»‰","chiáº¿c","cho","chá»©","chÆ°a","chuyá»‡n","cÃ³","cÃ³_thá»ƒ","cá»©","cá»§a","cÃ¹ng","cÅ©ng","Ä‘Ã£","Ä‘ang","Ä‘Ã¢y","Ä‘á»ƒ","Ä‘áº¿n_ná»—i","Ä‘á»u","Ä‘iá»u","do","Ä‘Ã³","Ä‘Æ°á»£c","dÆ°á»›i","gÃ¬","khi","khÃ´ng","lÃ "
,"láº¡i","lÃªn","lÃºc","mÃ ","má»—i","má»™t_cÃ¡ch","nÃ y","nÃªn","náº¿u","ngay","nhiá»u","nhÆ°","nhÆ°ng","nhá»¯ng","nÆ¡i","ná»¯a","pháº£i","qua","ra","ráº±ng","ráº±ng","ráº¥t","ráº¥t","rá»“i","sau","sáº½","so","sá»±","táº¡i","theo","thÃ¬","trÃªn","trÆ°á»›c"
,"tá»«","tá»«ng","vÃ ","váº«n","vÃ o","váº­y","vÃ¬","viá»‡c","vá»›i","vá»«a","trong","hÆ¡n","thá»ƒ","báº¡n","tÃ´i","ngÆ°á»i","mÃ¬nh","má»™t","cÃ²n","áº¥y","Ä‘áº¿n","á»Ÿ","báº¡n","vá»"}
for i in stop_word:
    all_text = all_text.replace(i, "")
print(all_text)
```

```python
# wordcloud táº§n suáº¥t cÃ¡c tá»« xuáº¥t hiá»‡n trong táº¥t cáº£ cÃ¡c post sau khi Ä‘Ã£ xÃ³a Ä‘i cÃ¡c stopwords viá»‡t nam 
from wordcloud import WordCloud, STOPWORDS
wordcloud = WordCloud(stopwords = STOPWORDS,background_color = 'black', max_words=300,width = 2000,height= 1200).generate(all_text)
plt.figure(figsize=(40,20))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# loáº¡i bá» cÃ¡c emoji khá»i all_text
import re
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
all_text = remove_emojis(all_text)
print(all_text)
```

```python
# chia all text thÃ nh list cÃ¡c tá»« riÃªng
text_list = all_text.split()
text_list
```

```python
# nhá»¯ng tá»« khÃ³a xuáº¥t hiá»‡n nhiá»u nháº¥t trong text Ä‘i kÃ¨m vá»›i sá»‘ lÆ°á»£t xuáº¥t hiá»‡n
from collections import Counter
Counter = Counter(text_list)
most_fre = Counter.most_common(30)
most_fre
```

```python
# chuyá»ƒn most_fre trÃªn thÃ nh dataframe Ä‘á»ƒ váº½ biá»ƒu Ä‘á»“ táº§n suáº¥t cÃ¡c keyword
word_frame = pd.DataFrame(most_fre, columns = ["KEYWORD", "FREQUENCY"])
word_frame
```

```python
# biá»ƒu Ä‘á»“ thá»ƒ hiá»‡n táº§n suáº¥t xuáº¥t hiá»‡n cá»§a cÃ¡c keyword xuáº¥t hiá»‡n nhiá»u nháº¥t
word_frame.plot.bar(x = 'KEYWORD',figsize = (8,6))
```

```python
# láº¥y ra list cÃ¡c key word xuáº¥t hiá»‡n nhiá»u nháº¥t
list_word = []
for i in range(0,30):
    list_word.append(most_fre[i][0])
list_word
```

```python
# xÃ³a bá»›t cÃ¡c tá»« khÃ´ng pháº£i lÃ  keyword 
for word in list_word:
    if word == '1' or word == '2' or word == 'vá»' or word == 'Of' or word == 'o' or word == 'The' or word == 'm' or word == 'ngÃ y':
        list_word.remove(word)
list_word
```

```python
post_imp = raw_df[['post_id','post_text','comments','shares','reaction_count']]
post_imp
```

```python
# tÃ¬m xem nhá»¯ng tá»« khÃ³a xuáº¥t hiá»‡n nhiá»u nháº¥t cÃ³ áº£nh hÆ°á»Ÿng Ä‘áº¿n lÆ°á»£t tÆ°Æ¡ng tÃ¡c bÃ i post khÃ´ng
# láº¥y theo thá»© tá»± giáº£m dáº§n theo reaction count
post_imp_order = post_imp.sort_values('reaction_count', ascending = False) 
post_imp_order
```

```python
# tÃ¬m sá»‘ láº§n xuáº¥t hiá»‡n cá»§a cÃ¡c keyword trong cÃ¡c bÃ i Ä‘Æ°á»£c nhiá»u lÆ°á»£t tÆ°Æ¡ng tÃ¡c nháº¥t
# Ä‘áº§u tiÃªn táº¡o dict gá»“m cÃ¡c key lÃ  keyword vÃ  value tÆ°Æ¡ng á»©ng lÃ  sá»‘ láº§n xuáº¥t hiá»‡n trong 20 bÃ i cÃ³ sá»‘ lÆ°á»£t tÆ°Æ¡ng tÃ¡c lá»›n nháº¥t
dict = {}
for text in post_imp_order['post_text'].head(20):
    for word in list_word:
        if word in text:
            dict[word] = 0
for text in post_imp_order['post_text'].head(20):
    for word in list_word:
        if word in text:
            dict[word] += 1
dict
```

```python
# convert tá»« dict sang dataframe Ä‘á»ƒ plot
x = dict.keys()
y = dict.values()
z = pd.DataFrame(x, columns= ['KEYWORD'])
z['FREQUENCY'] = pd.DataFrame(y)
z # lÃ  dataframe gá»“m keyword vÃ  sá»‘ láº§n xhien cá»§a nÃ³
```

```python
# sáº¯p xáº¿p giáº£m dáº§n cho dá»… nhÃ¬n tháº¥y word nÃ o xuáº¥t hiá»‡n nhiá»u nháº¥t 
z_order = z.sort_values('FREQUENCY', ascending= False)
z_order.plot.bar(x = 'KEYWORD',figsize = (8,6))
```
