```python
# lo·∫°i b·ªè c√°c k√≠ t·ª± ƒë·∫∑c bi·ªát, d·∫•u c√¢u
import string
map_table = str.maketrans("", "", string.punctuation)
all_text = all_text.translate(map_table)
print(all_text)
    
```

```python
# lo·∫°i b·ªè c√°c stopword trong all_text
stop_word = {"b·ªã","b·ªüi","c·∫£","c√°c","c√°i","c·∫ßn","c√†ng","ch·ªâ","chi·∫øc","cho","ch·ª©","ch∆∞a","chuy·ªán","c√≥","c√≥_th·ªÉ","c·ª©","c·ªßa","c√πng","c≈©ng","ƒë√£","ƒëang","ƒë√¢y","ƒë·ªÉ","ƒë·∫øn_n·ªói","ƒë·ªÅu","ƒëi·ªÅu","do","ƒë√≥","ƒë∆∞·ª£c","d∆∞·ªõi","g√¨","khi","kh√¥ng","l√†"
,"l·∫°i","l√™n","l√∫c","m√†","m·ªói","m·ªôt_c√°ch","n√†y","n√™n","n·∫øu","ngay","nhi·ªÅu","nh∆∞","nh∆∞ng","nh·ªØng","n∆°i","n·ªØa","ph·∫£i","qua","ra","r·∫±ng","r·∫±ng","r·∫•t","r·∫•t","r·ªìi","sau","s·∫Ω","so","s·ª±","t·∫°i","theo","th√¨","tr√™n","tr∆∞·ªõc"
,"t·ª´","t·ª´ng","v√†","v·∫´n","v√†o","v·∫≠y","v√¨","vi·ªác","v·ªõi","v·ª´a","trong","h∆°n","th·ªÉ","b·∫°n","t√¥i","ng∆∞·ªùi","m√¨nh","m·ªôt","c√≤n","·∫•y","ƒë·∫øn","·ªü","b·∫°n","v·ªÅ"}
for i in stop_word:
    all_text = all_text.replace(i, "")
print(all_text)
```

```python
# wordcloud t·∫ßn su·∫•t c√°c t·ª´ xu·∫•t hi·ªán trong t·∫•t c·∫£ c√°c post sau khi ƒë√£ x√≥a ƒëi c√°c stopwords vi·ªát nam 
from wordcloud import WordCloud, STOPWORDS
wordcloud = WordCloud(stopwords = STOPWORDS,background_color = 'black', max_words=300,width = 2000,height= 1200).generate(all_text)
plt.figure(figsize=(40,20))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# lo·∫°i b·ªè c√°c emoji kh·ªèi all_text
import re
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
all_text = remove_emojis(all_text)
print(all_text)
```

```python
# chia all text th√†nh list c√°c t·ª´ ri√™ng
text_list = all_text.split()
text_list
```

```python
# nh·ªØng t·ª´ kh√≥a xu·∫•t hi·ªán nhi·ªÅu nh·∫•t trong text ƒëi k√®m v·ªõi s·ªë l∆∞·ª£t xu·∫•t hi·ªán
from collections import Counter
Counter = Counter(text_list)
most_fre = Counter.most_common(30)
most_fre
```

```python
# chuy·ªÉn most_fre tr√™n th√†nh dataframe ƒë·ªÉ v·∫Ω bi·ªÉu ƒë·ªì t·∫ßn su·∫•t c√°c keyword
word_frame = pd.DataFrame(most_fre, columns = ["KEYWORD", "FREQUENCY"])
word_frame
```

```python
# bi·ªÉu ƒë·ªì th·ªÉ hi·ªán t·∫ßn su·∫•t xu·∫•t hi·ªán c·ªßa c√°c keyword xu·∫•t hi·ªán nhi·ªÅu nh·∫•t
word_frame.plot.bar(x = 'KEYWORD',figsize = (8,6))
```

```python
# l·∫•y ra list c√°c key word xu·∫•t hi·ªán nhi·ªÅu nh·∫•t
list_word = []
for i in range(0,30):
    list_word.append(most_fre[i][0])
list_word
```

```python
# x√≥a b·ªõt c√°c t·ª´ kh√¥ng ph·∫£i l√† keyword 
for word in list_word:
    if word == '1' or word == '2' or word == 'v·ªÅ' or word == 'Of' or word == 'o' or word == 'The' or word == 'm' or word == 'ng√†y':
        list_word.remove(word)
list_word
```

```python
post_imp = raw_df[['post_id','post_text','comments','shares','reaction_count']]
post_imp
```

```python
# t√¨m xem nh·ªØng t·ª´ kh√≥a xu·∫•t hi·ªán nhi·ªÅu nh·∫•t c√≥ ·∫£nh h∆∞·ªüng ƒë·∫øn l∆∞·ª£t t∆∞∆°ng t√°c b√†i post kh√¥ng
# l·∫•y theo th·ª© t·ª± gi·∫£m d·∫ßn theo reaction count
post_imp_order = post_imp.sort_values('reaction_count', ascending = False) 
post_imp_order
```

```python
# t√¨m s·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa c√°c keyword trong c√°c b√†i ƒë∆∞·ª£c nhi·ªÅu l∆∞·ª£t t∆∞∆°ng t√°c nh·∫•t
# ƒë·∫ßu ti√™n t·∫°o dict g·ªìm c√°c key l√† keyword v√† value t∆∞∆°ng ·ª©ng l√† s·ªë l·∫ßn xu·∫•t hi·ªán trong 20 b√†i c√≥ s·ªë l∆∞·ª£t t∆∞∆°ng t√°c l·ªõn nh·∫•t
dict = {}
for text in post_imp_order['post_text'].head(20):
    for word in list_word:
        if word in text:
            dict[word] = 0
for text in post_imp_order['post_text'].head(20):
    for word in list_word:
        if word in text:
            dict[word] += 1
dict
```

```python
# convert t·ª´ dict sang dataframe ƒë·ªÉ plot
x = dict.keys()
y = dict.values()
z = pd.DataFrame(x, columns= ['KEYWORD'])
z['FREQUENCY'] = pd.DataFrame(y)
z # l√† dataframe g·ªìm keyword v√† s·ªë l·∫ßn xhien c·ªßa n√≥
```

```python
# s·∫Øp x·∫øp gi·∫£m d·∫ßn cho d·ªÖ nh√¨n th·∫•y word n√†o xu·∫•t hi·ªán nhi·ªÅu nh·∫•t 
z_order = z.sort_values('FREQUENCY', ascending= False)
z_order.plot.bar(x = 'KEYWORD',figsize = (8,6))
```
