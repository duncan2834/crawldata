```python
# l·∫•y dataframe g·ªìm c√°c l∆∞·ª£t t∆∞∆°ng t√°c
post_react = raw_df[['post_text','reactions','shares', 'comments' ,'reaction_count']]
post_react
```

```python
# l·∫•y c√°c gi√° tr·ªã t·ª´ reactions ƒë·ªÉ t·∫°o th√†nh dataframe m·ªõi g·ªìm c√°c lo·∫°i react 
post_react['reactions'] = post_react['reactions'].apply(lambda x : dict(eval(x)))
post_all_react = post_react['reactions'].apply(pd.Series)
post_all_react
```

```python
# chuy·ªÉn c√°c gi√° tr·ªã NaN v·ªÅ 0 
post_all_react = post_all_react[['like','love','haha','wow','care','sad','angry']].fillna(0)
post_all_react
```

```python
# bi·ªÉu ƒë·ªì th·ªÉ hi·ªán c√°c lo·∫°i reaction theo t·ª´ng post
post_all_react.plot.line(figsize = (8, 6))
```

### TA TH·∫§Y L∆Ø·ª¢T LIKE V√Ä LOVE N·ªîI TR·ªòI H∆†N, V·ªÄ CU·ªêI   TH√å L∆Ø·ª¢T HAHA C√ì PH·∫¶N TƒÇNG V·ªåT

```python
post_imp = raw_df[['post_id','post_text','time','timestamp','likes','comments','shares','reaction_count']]
post_imp
```

```python
# b√†i vi·∫øt c√≥ nhi·ªÅu l∆∞·ª£t react nh·∫•t
post_max_react_count = post_imp[post_imp.reaction_count == post_imp.reaction_count.max()]
post_max_react_count
```

```python
#b√†i vi·∫øt c√≥ nhi·ªÅu l∆∞·ª£t comment nh·∫•t
post_max_comments = post_imp[post_imp.comments == post_imp.comments.max()]
post_max_comments
```

```python
# b√†i vi·∫øt c√≥ nhi·ªÅu l∆∞·ª£t share nh·∫•t
post_max_share = post_imp[post_imp.shares == post_imp.shares.max()]
post_max_share
```

```python
# l·∫•y gi√° tr·ªã gi·ªù trong tr∆∞·ªùng time ƒë·ªÉ l·∫•y ƒë∆∞·ª£c m·ªëc th·ªùi gian trong ng√†y
post_imp['onlytime'] = pd.to_datetime(post_imp['time'])
post_imp['onlytime'] = post_imp['onlytime'].dt.hour
post_time = post_imp[['post_text','reaction_count','shares','comments','onlytime']]
post_time
```

```python
# bi·ªÉu ƒë·ªì th·ªÉ hi·ªán c√°c m·ªëc th·ªùi gian trong ng√†y m√† page up post
post_time['onlytime'].value_counts().plot(kind = 'bar')
```

### Page th∆∞·ªùng up v√†o ƒë·∫ßu gi·ªù chi·ªÅu tr·ªü ƒëi cho ƒë·∫øn 22 gi·ªù t·ªëi

```python
# bi·ªÉu ƒë·ªì th·ªÉ hi·ªán s·ª± t∆∞∆°ng quan gi·ªØa s·ªë l∆∞·ª£ng reaction v√† comment v√† shares
post_share_cmt_react = post_imp[['comments', 'reaction_count','shares']]
post_share_cmt_react.plot.line(figsize = (8, 6))
```

### L∆Ø·ª¢NG REACTION L·ªöN H∆†N C·∫¢, SAU ƒê√ì ƒê·∫æN L∆Ø·ª¢T COMMENT V√Ä √çT NH·∫§T L√Ä L∆Ø·ª¢T SHARE

```python
# th√™m c·ªôt ƒë·ªô d√†i b√†i vi·∫øt 
post_reaction = raw_df[['post_text','reactions','shares', 'comments' ,'reaction_count']]
post_reaction['textlength'] = post_reaction['post_text'].str.len()
post_reaction
```

```python
# bi·ªÉu ƒë·ªì th·ªÉ hi·ªán s·ª± t∆∞∆°ng quan gi·ªØa l∆∞·ª£ng react v√† ƒë·ªô d√†i b√†i vi·∫øt
post_reaction[['reaction_count', 'textlength']].plot(kind = 'line', figsize=(8,6))
```

#### Ta th·∫•y ƒë·ªô d√†i b√†i vi·∫øt kh√¥ng ·∫£nh h∆∞·ªüng nhi·ªÅu ƒë·∫øn l∆∞·ª£ng react

```python
# l·∫•y t·∫•t c·∫£ c√°c post text 
all_text = ""
for i in post_imp['post_text']:
    all_text += str(i)
print(all_text)
```

```python
# wordcloud t·∫ßn su·∫•t c√°c t·ª´ xu·∫•t hi·ªán trong t·∫•t c·∫£ c√°c post
from wordcloud import WordCloud, STOPWORDS
wordcloud = WordCloud(stopwords = STOPWORDS,background_color = 'black', max_words=300,width = 2000,height= 1200).generate(all_text)
plt.figure(figsize=(40,20))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
